```python
import seaborn as sn
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
a=sn.load_dataset('car_crashes')
a.head(20)
```

**Now, we will plot a graph of abbrev against other attributes.**

```python
df=a.head(20)
total_states=sn.barplot(x="abbrev",y="total", data=df)
```

**In the above graph, we can see that AR has the highest accident rate and DC has the lowest accident rate.**

```python
alcohol_states=sn.barplot(x="abbrev",y="alcohol",data=df)
```

**In the above graph, HI state has highest alcohol related accidents rate and DC has the lowest alcohol related accidents rate.**

```python
ins_losses_states=sn.barplot(x="abbrev",y="ins_losses",data=df)
```

**In the above graph, LA has highest rate of loss for insurance companies per insured driver and ID has lowest loss rate.**

```python
ins_premium_state=sn.barplot(x="abbrev",y="ins_premium",data=df)
```

**In the above graph, LA and DC pays very high premiums while ID pays very low premiums.**

```python
speeding_states=sn.barplot(x="abbrev",y="speeding",data=df)
```

**From the above graph, we can say that HI has the highest accidents rate due to speeding and DC has the lowest accidents rate due to speeding.**

```python
no_previous_states=sn.barplot(x="abbrev",y="no_previous",data=df)
```

**From the above graph, AR has the highest accidents rate among people who never had accidents before and DC has the lowest accidents rate.**

```python
not_distracted_states=sn.barplot(x="abbrev",y="not_distracted",data=df)
```

**From the above graph, we can interpret that AR has the highest accidents rate among people who are not distracted while driving and DC has the lowest accidents rate**
